#Matplotlib in Python


*   Used to make graphs and visual representations of data in python



```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
print(x)
print('____________________________________________________________________________________')
print(y)
print('____________________________________________________________________________________')
print(z)
```

```python
# ploting the data
#sin wave
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sin value')
plt.title('SIN WAVE')
plt.show()
#cos wave
plt.plot(x,z)
plt.xlabel('angle')
plt.ylabel('cos value')
plt.title('COS WAVE')
plt.show()
```

```python
#making a Parabola Curve
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('value')
plt.title('Parabola Curve')
plt.show()
# making different forms of the same graph
print('Diffirent formating of the same graph')
plt.plot(x,y,'r+')
plt.show()

plt.plot(x,y,'g.')
plt.show()

plt.plot(x,y,'bX')
plt.show()
```

Multiple graphs in the same ploting space

```python
x = np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'r-')
plt.plot(x,np.cos(x),'g--')
plt.title('SIN & COS Graph')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

BAR Graphs

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
games = ['GTA San Andreas','Max Payne','Abooba the Alien','Monkey Game']
ranking = [10,9,7,6]
ax.bar(games,ranking)
plt.xlabel('Games')
plt.ylabel('Ranking of the games')
plt.title('Best game Graph')
plt.show()
```

PIE Chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
games = ['GTA San Andreas','Max Payne','Abooba the Alien','Monkey Game']
ranking = [10,9,7,6]
ax.pie(ranking,labels = games, autopct = '%1.1f%%')
plt.title('Best game Graph')
plt.show()
```

Scatter Plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='r')
ax.scatter(x,z,color='b')
plt.show()
```

3D Scatter Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection ='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,cmap='Blues')
plt.show()
```

#Seaborn in Python


*   Data Visualization Library



```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Seaborn has built in datasets

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# ploting/ploting the data
sns.relplot(data = tips,x = 'total_bill', y = 'tip', col='time', hue = 'smoker', style = 'smoker', size = 'size')
```

```python
# we can also set themes for our graphs
sns.set_theme()
sns.relplot(data = tips,x = 'total_bill', y = 'tip', col='time', hue = 'smoker', style = 'smoker', size = 'size')
```
